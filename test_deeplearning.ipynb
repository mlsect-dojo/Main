

Ниже приведены задачи тестового задания по глубокому обучению. Материал заданий покрывается курсом от Samsung AI Center. Другие курсы по DL тоже должны покрывать темы задания. Вам нужно выполнить задания до дедлайна, защитить результаты и быть очень гордыми и довольными тем, какие вы умнички

# Задачи

## 1. Начнём с простого

Обучите полносвязную сеть для классификации съедобности грибов ([Данные](https://www.kaggle.com/uciml/mushroom-classification))

Допускается совместить классы "съедобность неизвестна" и "несъедобно"

Оформление решений:
- Используйте гитхаб. Это нужно и для отката ваших локальных изменений, и для удобства проверки - он рендерит ноутбуки при просмотре в браузере
- Для каждой задачи создайте отдельную папку. В папке должен быть ноутбук с тренировкой модели и отчётом о результатах (с графиками, цифрами, выводами и картинкой с ПСЖ)
- Используйте .gitignore. Не загружайте данные на гит
- Желательно использовать torch. Но не страшно, если это будет другой фреймворк для глубокого обучения
- Не крадите чужой код. Тестовое задание создано для закрепления ваших знаний и навыков, а не для галочки внутри комьюнити. Всё сделанное понадобится вам на дальнешем пути
- Задавайте вопросы в чате, если вам что-либо непонятно
- Не откладывайте выполнение задания на потом

## 2. Сегментация

Обучить нейросеть для сегментации рака головного мозга ([Данные](https://www.kaggle.com/mateuszbuda/lgg-mri-segmentation))

Подробно:

- Считайте изображения МРТ и соответствующие им карты сегментации. Данные о картинках находятся в csv файле
- Примените аугментации к изображениям. [Подбробнее о них](https://medium.com/@dc.aihub/data-augmentation-for-neural-networks-and-computer-vision-b4f993c34e91)
- Разделите данные на train и test выборки
- Сгруппируйте данные по батчам
- Реализуйте модель для сегментации. ([Абстрактно об архитектурах для CV задач](https://towardsdatascience.com/how-to-do-everything-in-computer-vision-2b442c469928)), ([Общее описание популярных архитектур для сегментации картинок](https://medium.com/swlh/image-segmentation-using-deep-learning-a-survey-e37e0f0a1489))
- Напишите цикл обучения модели - вычисление метрик, ошибки, шаг оптимизации параметров сети
- \*Для визуализации метрик можно использовать tensorboard. Его легко добавить в код, он поможет быстрее и проще понимать, что вы учите и как успешно вы это делаете. Добавляйте в него информацию о метриках и примеры работы сети - тройки картинок (оригинал, настоящая маска с опухолью, предсказанная маска). Это значительно улучшит понимание происходящего
- Подберите параметры для модели, максимизирующие качество её предсказаний на test выборке. Проверьте, что сеть не переобучена на train выборку
- Выберите другую архитектуру модели, разберитесь в ней. Найдите реализацию модели на гитхабе и обучите на своих данных
- Напишите отчёт о решении задачи. В нём обязательно должны быть: 
  1. описание выбранной архитектуры сети
  2. описание функции ошибки
  3. графики ошибок на train и test
  4. несколько примеров предсказаний сети. Пример предсказания - тройка картинок (оригинал, настоящая маска, предсказанная маска)
  5. выводы о качестве модели
  6. пункты 1-5 для модели с гитхаба
  7. сравнение моделей
  8. \*по желанию: выводы по задаче, улучшению своего/чужого решения, что-то ещё
  

Советы:
- Не загружайте все картинки в память. Храните пути, а при обращении к объекту датасета загружайте нужное изображение
- Для хранения датасета и деления на батчи используйте классы Dataset и Dataloader (если вы torch (или аналоги, если вы не т0рч))
- Прочитайте несколько туториалов на medium или towardsdatascience о выбранной модели. Разберитесь в принципах её работы
- Если у вас не получается реализовать часть модели или непонятна какая-то её часть - не бойтесь задавать вопросы в конфе
- Для работы с tensorboard в torch есть библиотека tensorboardX
- Для удобного просмотра слоёв модели и размера возвращаемых из них картинок есть библиотека torchsummaryX (тоже для torch. Для tensorflow удобнее выводить граф модели в tensorboard)
- Для аугментаций можно(нужно) использовать преобразования из torchvision.transforms. Там же есть resize, используйте его, потому что сеть для картинок разрешения больше 200х200 будет учиться долго. Особенно если вы не можете использовать GPU. В таких случаях можете попробовать Google Colab

- Подберите параметры для модели, максимизирующие качество её предсказаний на test выборке. Проверьте, что сеть не переобучена на train выборку
- Напишите отчёт о решении задачи. В нём обязательно должны быть: 
  1. описание выбранной архитектуры сети
  2. описание функции ошибки

  4. несколько примеров предсказаний сети. Пример предсказания - тройка картинок (оригинал, настоящая маска, предсказанная маска)
  5. выводы о качестве модели
  8. \*по желанию: выводы по задаче, улучшению своего решения, что-то ещё
